# Iris Dataset Exploration

Before writing train.py I wanted to understand the dataset first — class distribution, feature ranges, whether the classes are easy to separate. This notebook is that exploration.

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

print("Shape:", df.shape)
df.head()
```

## Class distribution

Checking if the classes are balanced — imbalanced classes would affect how I interpret the F1 score.

```python
df['species'].value_counts()
```

## Feature ranges

Checking the range of each feature across species. If the ranges overlap heavily the classifier will have a harder job.

```python
df.groupby('species').agg(['min', 'max', 'mean']).round(2)
```

## Notes after exploration

- All three classes have exactly 50 samples each — perfectly balanced, so accuracy and F1 will be very close to each other on this dataset.
- Petal length and petal width separate the classes much more cleanly than sepal measurements. Setosa is completely separate. Versicolor and virginica overlap slightly.
- This explains why RandomForest gets perfect or near-perfect scores — the data is clean and well-separated. On a real dataset with noise and missing values the scores would be lower and the threshold choice would matter more.
- In my thesis data (DHS health survey) there was significant missingness and class imbalance — very different from this. That context is useful to keep in mind when reading the pipeline results here.
